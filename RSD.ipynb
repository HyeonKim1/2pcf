```python
import camb
from camb import model, initialpower
import numpy as np
from matplotlib import pyplot as plt
```

```python
nn=100
n=50
redshift=np.linspace(0,1,nn)
H0=np.linspace(60,80,n)

sigma8_vals=np.zeros((n,nn))
for ii,jj in enumerate(H0):
    pars = camb.CAMBparams()
    pars.set_cosmology(H0=jj, ombh2=0.022, omch2=0.122)
    pars.set_matter_power(redshifts=redshift)
    pars.NonLinear = model.NonLinear_none  # 선형 전력 스펙트럼 사용

    results = camb.get_results(pars)

    sigma8_vals[ii] = results.get_fsigma8()  # 각 z에 대한 sigma8 배열
```

```python
fig, ax = plt.subplots(figsize=(7, 3))

cmap = plt.cm.viridis
norm = plt.Normalize(vmin=H0.min(), vmax=H0.max())


for ii in range(n):
    ax.plot(redshift, sigma8_vals[ii], c=cmap(norm(H0[ii])))


sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

cbar = fig.colorbar(sm, ax=ax, label=r'$H_0$')

ax.set_xlabel('z')
ax.set_ylabel(r'$f\sigma_8$')

plt.show()
```
